```python
from models.utils import loadData

x, y = loadData()
```

```python
print("Input features: ", x)
```

```python
print("Output Labels: ", y)
```

```python
from models.utils import MinMaxScaler

scaler = MinMaxScaler()
xNorm = scaler.fitTransform(x)
print(xNorm)
```

```python
from models.utils import LinearRegression

model = LinearRegression()
print(model.__dict__)
model.fit(xNorm, y, epochs=50, a=0.5)
yHat = model.predict(10)
print(yHat)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.plot(list(range(1, 51)), model.costs)
ax.set_title("Cost vs Epoch")
ax.set_xlabel("Epoch")
ax.set_ylabel("Cost")
plt.show()
```

```python
_, ax = plt.subplots()
ax.scatter(x, y)
ax.set_title("Dataset")
ax.set_xlabel("Features")
ax.set_ylabel("Labels")
plt.show()
```

```python
model = LinearRegression()
model.fit(xNorm, y, epochs=50, a=0.005)
yHat = model.predict(10)
print(yHat)
```
